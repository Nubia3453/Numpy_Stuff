```python
# import requred pakages
import numpy as np

# get the data from source file to python code
data=np.genfromtxt(r"C:\Users\Vinayak\Downloads\nyc_taxis.csv", delimiter=",", skip_header=True)

#calculate speed
#speed=distance(km)/time(hr)

speed=data[:,7]/(data[:,8]/3600)

#calculating mean speed.
mean_speed=speed.mean()
print('Mean speed :',mean_speed,'km/hr')
```

```python
#determining total rides in month of feb
#using operator logic to spaperate out requred column only
feb_rides=data[data[:, 1]==2,1]
print('Total rides in Feb:',feb_rides.shape[0])
```

```python
#trips with tip greater than $50.
tip=data[data[:,12]>50,1]
tip.shape[0]
print('There are',tip.shape[0], 'trips where tip offred is more than 50 bucks')
```

```python
#no of drops happended on JFK airport
#location code for JFK airport is 2
drop_at_jfk=data[data[:,6]==2]
drop_at_jfk.shape[0]
print(drop_at_jfk.shape[0], 'drops happened on JFK airport')
```

```python
# calculate no of payments made by cash and calculate total amount
# payment code for cash paytmets is 1
pay_by_cash=data[data[:,-1]==1]
pay_by_cash.shape[0]

print("No of payments made by cash is:",pay_by_cash.shape[0])

#calculate total amount

print("toatl", pay_by_cash.sum(), 'dollers of payment made by cash')
```
